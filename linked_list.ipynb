```python
# Node class 
class Node:   
    def __init__(self, data): 
        self.data = data  # Assign data 
        self.value = data  # Assign data 
        self.next = None  # Initialize next as null 
        # doubly linked list
        #self.previous = None

# Linked List class contains a Node object 
class LinkedList:   
    # Function to initialize head 
    def __init__(self): 
        self.head = None
        
    # print the linked list
    def print_list(self):
        current_node = self.head 
        while (current_node): 
            print (current_node.data) 
            current_node = current_node.next

    # insert data at beginning 
    def add_data_at_beginning(self,new_data):
        # allocate new_data to a Node class
        new_node = Node(new_data)
        # next value is previous list head 
        new_node.next = self.head
        # head points to new_node
        self.head = new_node        
            
    def add_data_at_end(self, new_data):
        # allocate new_data to a Node class
        new_node = Node(new_data) 
        # if the list is empty, make the new_node the head 
        if self.head is None: 
            self.head = new_node 
            return
        # traverse the full list        
        last = self.head
        while(last.next):
            last = last.next
        # insert new node at the end 
        last.next = new_node
 
    # Function to add node
    def add_data_after_middle_node(self, middle_node, new_data):
        # allocate new_data to a Node class
        new_node = Node(new_data)
        #middle_node = Node(middle_data)
        # check that middle_node exists
        if middle_node is None:
            print("The mentioned node is absent")
            return          
        # next of new_node is next of middle_node 
        new_node.next = middle_node.next
        # next of middle_node is new_node 
        middle_node.next = new_node 
        
    def delete_node(self, data_to_delete):  
        # start at the head 
        current_node = self.head  
        previous_node = None
        found = False
        while (current_node and not found): 
            if (current_node.data == data_to_delete):  
                found = True 
            else:
                previous_node = current_node
                current_node = current_node.next
        # if key was not present in linked list  
        if(current_node == None):
            print('key is not present in linked list')
            return
        if (previous_node is None):
            self.head = current_node.next
        else:
            # Unlink the node from linked list  
            previous_node.next = current_node.next  
            current_node = None
        
    def count_length_of_list(self):            
        current_node = self.head  
        # If head node itself holds the key to be deleted  
        list_length = 0
        while (current_node): 
            current_node = current_node.next
            list_length += 1        
        return list_length 
    
```
