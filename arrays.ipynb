```python
class Solution:
    
    def subarraysDivByK(self, nums: List[int], k: int) -> int:
        cache = dict()
        cache[0] = 1
        count = 0
        run_sum = 0
        for i in range(len(nums)):
            run_sum += nums[i]
            if run_sum%k in cache:
                count += cache[run_sum%k]
                cache[run_sum%k] += 1
            else:
                cache[run_sum%k] = 1

        return count
    
    def gridIllumination(self, N, lamps, queries):
        dirs = [[-1,-1], [-1,0], [-1,1], [0,-1], [0,0], [0,1], [1,-1], [1,0], [1,1]]
        light_set = set((y,x) for y,x in lamps)
        horizontal = collections.defaultdict(int)
        vertical   = collections.defaultdict(int)
        diag       = collections.defaultdict(int)
        anti_diag  = collections.defaultdict(int)        
        for j, i in lamps:
            horizontal[j] += 1
            vertical[i] += 1
            diag[j-i] +=1 
            anti_diag[j+i] +=1 
        result = []
        for j, i in queries:
            if horizontal[j] > 0 or vertical[i] > 0 or diag[j-i] > 0 or anti_diag[j+i] > 0:
                result.append(1)
            else:
                result.append(0)
            for dir in dirs:
                jj, ii = j+dir[0], i+dir[1]
                if (jj, ii) in light_set:
                    light_set.remove((jj, ii))
                    horizontal[jj] -= 1
                    vertical[ii] -= 1
                    diag[jj-ii] -= 1
                    anti_diag[jj+ii] -= 1
        
        return result
    
    def isGoodArray(self, nums): # bezouts theorem, pure memorization
        if not nums: return        
        # if gcd(a,b) = k there exists int x, y such a*x+b*y = k
        # if gcd(all integers in array) = 1, there exists some subset and x,y such that a*x+b*y = 1        
        def gcd(a,b):
            while b:
                a, b = b, a%b
            return a
        x = nums[0]
        for num in nums:
            x = gcd(x, num) 
        if x == 1: return True
        else: return False        

    def maxSlidingWindow(self, nums, k):
        # queue of possible indices, biggest number always on top, max length k 
        queue = collections.deque()
        result = []
        for index, num in enumerate(nums):
            # top of queue is always max of queue and current num 
            while queue and nums[queue[-1]] < num: 
                queue.pop()
            queue.append(index)
            if queue[0] == index-k: # queue is too long
                queue.popleft()
            if index >= k-1: # skip first k numbers 
                result.append(nums[queue[0]])
        return result
     
    def trap(self, height):
        # fill to height of the left
        # drain down to height of fill to the right 
        volume = []
        left = 0
        for ht in height:
            left = max(left, ht)
            volume.append(left)
        volume.reverse()
        height.reverse()
        right = 0
        total_water = 0
        for index, ht in enumerate(height):
            right = max(right, ht)
            total_water += min(volume[index], right) - ht
        return total_water    

    def trap(self, height):
        if not height or len(height) < 3:
            return 0
        volume = 0
        left, right = 0, len(height) - 1
        l_max, r_max = height[left], height[right]
        while left < right:
            l_max, r_max = max(height[left], l_max), max(height[right], r_max)
            if l_max <= r_max:
                volume += l_max - height[left]
                left += 1
            else:
                volume += r_max - height[right]
                right -= 1
        return volume
    
```
