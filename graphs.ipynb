```python

class Solution:

    def allPathsSourceTarget(self, graph): # dag
        graph_list = collections.defaultdict(list)
        for index, children in enumerate(graph):
            graph_list[index] = children
        result = []
        def dfs(path, node, target):
            stack = graph_list[node]
            if not stack:
                return
            for n in range(0, len(stack)):
                if stack[n] == target:
                    result.append(path+[stack[n]])
                else:
                    dfs(path+[stack[n]], stack[n], target)
        dfs([0],0,len(graph)-1)
        return result
    
    def killProcess(self, pid, ppid, kill):
        graph = defaultdict(list)
        stack = deque([kill])
        result = []
        for parent, child in zip(ppid, pid):
            graph[parent].append(child)
        while stack:
            parent = stack.pop()
            result.append(parent)
            stack.extend(graph[parent])
        return result
```
