```python
class Solution:
    def reverseString1(self, string):
        if len(string)<=1:
            return string
        n = len(string)
        return self.reverseString1(string[n//2:])+self.reverseString1(string[:n//2])  
    
    def reverseString2(self, string):
        r = list(string)
        i, j  = 0, len(r) - 1
        while i < j:
            r[i], r[j] = r[j], r[i]
            i += 1
            j -= 1

        return "".join(r)
    
    def reverseString3(self, s):
        len_str = len(s)
        string_list = list(s)
        for n in range(0, round(len_str/2)+1, 1):
            #print(sample_string[n])
            (string_list[n], string_list[len_str-n-1]) = (string_list[len_str-n-1], string_list[n])
        return "".join(string_list)
    
    
    
        
```
